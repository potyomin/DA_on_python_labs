# Задача 2

```python
df = pd.read_csv(FILE_BEVERAGE, sep=";", index_col="numb.obs")
df.head()
```

```python
np.random.seed(seed=2)
r1 = np.random.uniform(low=0.0, high=1.0, size=len(df))
```

```python
q1 = np.quantile(r1, 0.25)
q2 = np.quantile(r1, 0.50)
q3 = np.quantile(r1, 0.75)
```

```python
m1 = (r1 <= q1)
m2 = (r1 > q1) & (r1 <= q2)
m3 = (r1 > q2) & (r1 <= q3)
m4 = (r1 > q3)
```

```python
df_1 = df.iloc[m1, :]
df_2 = df.iloc[m2, :]
df_3 = df.iloc[m3, :]
df_4 = df.iloc[m4, :]
```

```python
print("Всего:", len(df))
print("Часть 1:", len(df_1))
print("Часть 2:", len(df_2))
print("Часть 3:", len(df_3))
print("Часть 4:", len(df_4))
```

# Задача 3

```python
df = pd.read_csv(FILE_BEVERAGE, sep=";", index_col="numb.obs")
```

```python
col_coke = "COKE"
col_tab  = "TAB"
```

```python
np.random.seed(seed=42)
```

```python
mask_part1 = np.zeros(len(df), dtype=bool)
```

```python
for v1 in [0, 1]:
    for v2 in [0, 1]:

        idx = (df[col_coke] == v1) & (df[col_tab] == v2)
        n = idx.sum()

        if n == 0:
            continue

        r = np.random.uniform(low=0.0, high=1.0, size=n)

        if n == 1:
            take = (r > 0.5)
        else:
            take = (r > np.quantile(r, 0.5))

        mask_part1[idx] = take
```

```python
df_1 = df.iloc[mask_part1, :]
df_2 = df.iloc[np.logical_not(mask_part1), :]
```

```python
df_1
```

```python
df_2
```

```python
print("Размер первой подвыборки: ", len(df_1))
print("Размер второй подвыборки: ", len(df_2))
```
